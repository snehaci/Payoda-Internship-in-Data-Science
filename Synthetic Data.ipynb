# SYNTHETIC DATA

#### Artificially made by computer programs

### why synthetic data?
#### 1) Lack of control over data
#### 2) Data Privacy

### Ways to generate Synthetic Data

### 1) SMOTE - Synthetic Minority Over-Sampling Technique
#####                 Helpful in classification problems where we have imbalance data

### 2) Data Augmentation
##### Helpful in image datasets 

### 3) Variational Auto Encoder
#### Encodes input data in gaussian distribution form

### 4) Cut and Paste
#### Create new data from previously available image data

### 5) GAN - Generative Adversarial Network
##### Discriminator Model (identifies the fake and real images)
##### Generator Model (create new fake images that closely relates real images)
